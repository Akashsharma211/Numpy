```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
```

```python
print("\n Sales data for 1st 3 restaurant\n", sales_data[:3])
```

```python
print("\n Sales data for 1st 3 restaurant\n", sales_data[:, 1:])
```

# total sale per year

```python
print(np.sum(sales_data, axis=0))

yearly_total = np.sum(sales_data[:, 1:], axis=0)
print("\n Yearly total sales from 2021 to 2024:", yearly_total)
```

```python
# minimum sales per annum
min_sales = np.min(sales_data[:, 1:], axis=0)
print("\n Minimum sales from 2021 to 2024:", min_sales)
```

```python
#maximum sales per annum
max_sales = np.max(sales_data[:, 1:], axis=0)
print("\n Maximum sales from 2021 to 2024:", max_sales)
```

```python
#Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("\n Average sales from 2021 to 2024:", avg_sales)
```

```python
cumulative_sales = np.cumsum(sales_data[:, 1:], axis=0)
print("\n Cumulative sales from 2021 to 2024:\n", cumulative_sales)
```

```python
cumulative_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("\n Cumulative sales from 2021 to 2024:\n", cumulative_sales)

plt.figure(figsize=(8,6))
plt.plot(np.mean(cumulative_sales, axis=0))
plt.title("Average Cumulative Sales across Restaurants (2021-2024)")
plt.xlabel("Years from 2021 to 2024")
plt.ylabel("Average Cumulative Sales")
plt.grid(True)
plt.show()
```

## VECTOR

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition:", vector1 + vector2)

print("Vector Multiplication:", vector1 * vector2)

print("Dot Product:", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle between vectors (radians):", angle)
```

```python
restaurant_types = np.array(['biryani', 'chinese', 'pizza', 'burger', 'cafe'])
vectorized_upper = np.vectorize(str.upper)
print(" Vectorized Uppercase Restaurant Types:", vectorized_upper(restaurant_types))
```
